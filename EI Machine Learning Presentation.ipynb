```python
# http://stackoverflow.com/questions/27298178/concatenate-strings-from-several-rows-using-pandas-groupby
idf['compiled_int'] = idf[['username','interest']].groupby(['username'])['interest'].transform(lambda x: ', '.join(x))
# del idf['interest']

# # http://stackoverflow.com/questions/23667369/drop-all-duplicate-rows-in-python-pandas
idf.drop_duplicates(subset=['username'], keep='last', inplace=True)
# print idf.shape
idf.head(10)
```

```python
result = pd.merge(cdf, idf[['username','compiled_int']], how='left', on=['username'])
result.head()
```

```python
# http://stackoverflow.com/questions/18658106/quick-implementation-of-character-n-grams-using-python
def get_ngrams(phr,n=3):
    #ensure length of word > n, else return entire word
    if len(phr)<=n:
        return [phr]
    ngrams = [phr[i:i+n] for i in range(len(phr)-n+1)]
#     print ngrams
    return ngrams

# http://stackoverflow.com/questions/22428020/how-to-extract-character-ngram-from-sentences-python
# import re
# def get_ngrams2(phr):
#     ngrams = re.findall(r'(?=(...))',phr)
#     print ngrams
```

```python
#create list of all features
def get_list(col, n=60):
    c=Counter()
    FEATURES_NUM = n
    for string in col:
        ngrams = get_ngrams(string)
        for i in ngrams:
            c[i] += 1
    words = [word for word, count in c.most_common(FEATURES_NUM)]
    # words = sorted(c, key = c.get, reverse = True)[:FEATURES_NUM] #alternative
    tuple_counts = [tup for tup in c.most_common(FEATURES_NUM)]
    return words, tuple_counts
```

```python
#making feature vector for every person
def get_vector(ngrams_user, ngrams_list):
    res_vector = []
    for k, ngram in enumerate(ngrams_list):
        res_vector.append(int(ngram in ngrams_user)) #convert boolean to int
    return res_vector

occ_ngrams, occ_counts = get_list(result.occupation)
print occ_ngrams
int_ngrams, int_counts = get_list(result.compiled_int)
print int_ngrams

# # test for one user
# ngram_user = get_ngrams("Consulting, Marketing, Strategic Management")
# arr = np.array(get_vector(ngram_user,int_ngrams)) 
# print arr

# get vectors for each user
user_vectors = [] # start with list, then convert to np array
vector = []
for i in range(len(result)):
    user_occ = get_vector(get_ngrams(result.occupation[i]),occ_ngrams)
    user_int = get_vector(get_ngrams(result.compiled_int[i]),int_ngrams)
    vector = np.concatenate((user_occ, user_int))
    user_vectors.append(vector)
user_vectors = np.asarray(user_vectors)

# print user_vectors[:3]
```

```python
# Perform k-Means clustering
from sklearn.cluster import KMeans

print('Vectors created; clustering is starting now!')
X = pd.DataFrame(user_vectors)
CLUSTERS_COUNT = len(cdf)/10

km = KMeans(n_clusters=CLUSTERS_COUNT, random_state=1)
km.fit(X)
km.labels_
X['cluster'] = km.labels_
X.sort('cluster')
combined = pd.concat([X[['cluster']], result.drop(['name'],axis=1)], axis=1)
combined.sort_values(by='cluster').head(10)
```
